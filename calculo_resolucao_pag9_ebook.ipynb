Questão: Pense em algo comum, vamos supor que você queira saber se a preferência por tons de batom (nude ou vermelho, por exemplo) muda entre pessoas de diferentes faixas etárias.

```python
'''importanto as bibliotecas necessarias'''

import pandas as pd
import scipy.stats as stats
```

```python
'''criando um data frame (df), com os dados da tabela apresentada no e-book'''

data={'Faixa Etária':['18-25 anos', '26-35 anos', 'Total'],
        'Nude':[40, 25, 65],
        'Vermelho':[20, 35, 55],
        'Total':[60, 60, 120]}
df=pd.DataFrame(data)
df=df.set_index('Faixa Etária')
```

```python
'''criando e visualizando a tabela de contigência'''

contingency_table=df.iloc[:2, :2]
df
```

```python
'''realizando o teste qui-quadrado, onde:
chi2_stat: valor do Qui-Quadrado calculado.
p_val: valor-p do teste.
dof: graus de liberdade.
expected: frequências esperadas em cada célula da tabela, caso não houvesse associação.'''

chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table, correction=False)
```

```python
'''definindo os parâmetros do teste neste caso:
os graus de liberdade (df), o nível de significância (alpha) e calcula o valor crítico para comparação.'''

df = 1  # Graus de liberdade
alpha = 0.05  # Nível de significância
critical_value = stats.chi2.ppf(1 - alpha, df)
```

```python
'''imprimindo os resultados'''

print(f"Valor do Qui-Quadrado: {chi2_stat}")
print(f"Valor-p: {p_val}")
print(f"Graus de liberdade: {dof}")
print(f"Graus de liberdade: {dof}")
print(f"Valor crítico: {critical_value}")
print("Frequências Esperadas:")
print(expected)
print("Frequências Esperadas:")
print(expected)
```
